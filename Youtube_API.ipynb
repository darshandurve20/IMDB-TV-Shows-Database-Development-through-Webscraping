# IMPORTING THE LIBRARIES

```python
import pandas as pd
from requests import get
from bs4 import BeautifulSoup as bs
```

# GETTING THE TV SHOWS

```python
url = "https://www.imdb.com/search/title?title_type=tv_series"
response = get(url)
html_soup = bs(response.text ,'html.parser')
id_check = html_soup.find(id ="main")
tv_show_container = id_check.find_all(class_ ="lister-item mode-advanced")
container = tv_show_container[0]
tv_show_names = []
for container in tv_show_container:

    if container.find("div", class_ = "ratings-bar") is not None:

        name = container.h3.a.text
        tv_show_names.append(name)

tv_show_df = pd.DataFrame({"TV Show Name" :tv_show_names})
tv_show_df
```

# BUILDING THE YOUTUBE API CALL

```python
api_key = 'AIzaSyBELQuUpv6O6vUt31QFoFrr7i6Zzymc5fA'

from apiclient.discovery import build

youtube = build('youtube', 'v3', developerKey = api_key)
```

# GETTING THE YOUTUBE VIDEO TITLE AND DESCRIPTION ALONG WITH THE VIDEO ID

```python
tv_show_name=[]
videoid=[]
videotitle=[]
desc=[]
tv_show=[]
video_id=[]
video_title=[]
description=[]
for index, row in tv_show_df.iterrows(): 
    req = youtube.search().list(q=row['TV Show Name'], part='snippet', type='video', maxResults=1)
    res = req.execute()

    for item in res['items']:
        tv_show.append(row['TV Show Name'])
        videoid.append(item['id']['videoId'])
        videotitle.append(item['snippet']['title'])
        desc.append(item['snippet']['description'])
    
    tv_show_name.extend(tv_show)
    video_id.extend(videoid)
    video_title.extend(videotitle)
    description.extend(desc)
    
    tv_show.clear()
    videoid.clear()
    videotitle.clear()
    desc.clear()

youtube_api_df_1=pd.DataFrame({'TV Show Name':tv_show_name,
                              'Video ID':video_id,
                              'Video title':video_title,
                              'Description':description})
youtube_api_df_1
```

# GETTING VIEWS, LIKES, DISLIKES AND VIDEO DURATION OF TV SHOWS VIDEOS

```python
video_Id=[]
view_Count=[]
like_Count=[]
dislike_Count=[]
video_Duration=[]

videoId=[]
viewCount=[]
likeCount=[]
dislikeCount=[]
videoDuration=[]

for index, row in youtube_api_df_1.iterrows():  

    
    req = youtube.videos().list(part='snippet,statistics,contentDetails', id=row['Video ID'])
    res = req.execute()

    for item in res['items']:
        videoId.append(row['Video ID'])
        try:
            viewCount.append(item['statistics']['viewCount'])
        except:
            viewCount.append(None)
        try:
            likeCount.append(item['statistics']['likeCount'])
        except:
            likeCount.append(None)
        try:
            dislikeCount.append(item['statistics']['dislikeCount'])
        except:
            dislikeCount.append(None)
        videoDuration.append(item['contentDetails']['duration'])
    
    video_Id.extend(videoId)
    view_Count.extend(viewCount)
    like_Count.extend(likeCount)
    dislike_Count.extend(dislikeCount)
    video_Duration.extend(videoDuration)

    videoId.clear()
    viewCount.clear()
    likeCount.clear()
    dislikeCount.clear()
    videoDuration.clear()

youtube_api_df_2=pd.DataFrame({'Video ID':video_Id,
                               'Views':view_Count,
                         'Likes':like_Count,
                         'Dislikes':dislike_Count,
                        'Video Duration':video_Duration})
youtube_api_df_2
```

# JOINING THE DATAFRAMES ON VIDEO ID TO FORM YOUTUBE API DATAFRAME

```python
youtube_api_df=pd.merge(youtube_api_df_1,youtube_api_df_2,on='Video ID')
youtube_api_df
```

# GETTING THE COMMENTS FOR THE VIDEOS OF THE TV SHOWS INTO YOUTUBE COMMENTS DATAFRAME

```python
comments=[]
vidid=[]
commentz=[]
videoid=[]

for index, row in youtube_api_df_1.iterrows():
    req = youtube.commentThreads().list(part='snippet', videoId=row['Video ID'], maxResults=5) #googleapiclient.http.HttpRequest

    res = req.execute()

    for item in res['items']:
        videoid.append(item['snippet']['videoId'])
        commentz.append(item['snippet']['topLevelComment']['snippet']['textOriginal'])
    comments.extend(commentz)
    vidid.extend(videoid)
    videoid.clear()
    commentz.clear()
    
comments_df=pd.DataFrame({'Video ID':vidid,
                        'comments':comments})
comments_df
```

# CONVERTING THE DATA FRAMES TO CSV FILES

```python
comments_df.to_csv('Youtube_Comments1.csv', index=False)
youtube_api_df.to_csv('Youtube_API1.csv', index=False)
```
